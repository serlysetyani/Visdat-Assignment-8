```python
import numpy as np
import pandas as pd
from os import path
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import re
import string

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/test.csv', index_col=0)
df.head()
```

```python
data_len = len(df)
print("Dataframe Length: ", data_len)
```

```python
print("There are {} observations and {} features in this dataset. \n".format(df.shape[0],df.shape[1]))
```

```python
df.to_numpy()
```

```python
list_review = []
for i in df['user_review']:
  list_review.append(i)
```

```python
wordcloud = WordCloud(width=1600, height=800, background_color="white").generate(''.join(str(list_review)))

plt.figure( figsize=(20,10) )
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
